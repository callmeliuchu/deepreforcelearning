```python
import flappy_bird_gymnasium
import gymnasium as gym
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical
import matplotlib.pyplot as plt

# 设置随机种子以确保结果可复现
torch.manual_seed(42)
np.random.seed(42)

# 创建环境
# env = gym.make("FlappyBird-v0", render_mode="human", use_lidar=True)
env = gym.make("FlappyBird-v0", use_lidar=False)
obs, info = env.reset()

# 定义策略网络
class PolicyNetwork(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(PolicyNetwork, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, output_dim),
            nn.Softmax(dim=-1)
        )
    
    def forward(self, x):
        return self.network(x)

# 策略梯度智能体
class PolicyGradientAgent:
    def __init__(self, input_dim, output_dim, learning_rate=0.001, gamma=0.99):
        self.policy = PolicyNetwork(input_dim, output_dim)
        self.optimizer = optim.Adam(self.policy.parameters(), lr=learning_rate)
        self.gamma = gamma
        
        self.log_probs = []
        self.rewards = []
        
    def select_action(self, state):
        state = torch.FloatTensor(state)
        probs = self.policy(state)
        dist = Categorical(probs)
        action = dist.sample()
        log_prob = dist.log_prob(action)
        self.log_probs.append(log_prob)
        return action.item()
    
    def store_reward(self, reward):
        self.rewards.append(reward)
    
    def learn(self):
        # 如果没有收集到任何经验，直接返回
        if len(self.rewards) == 0:
            return
            
        # 计算折扣回报
        returns = []
        R = 0
        for r in reversed(self.rewards):
            R = r + self.gamma * R
            returns.insert(0, R)
        
        # 标准化回报以减少方差
        returns = torch.tensor(returns)
        if len(returns) > 1:  # 只有当有多个回报时才标准化
            returns = (returns - returns.mean()) / (returns.std() + 1e-9)
        
        # 计算损失
        policy_loss = []
        for log_prob, R in zip(self.log_probs, returns):
            policy_loss.append(-log_prob * R)
        
        # 确保 policy_loss 不为空
        if len(policy_loss) > 0:
            # 使用 stack 而不是 cat 来处理零维张量
            if len(policy_loss) == 1:
                policy_loss = policy_loss[0]
            else:
                policy_loss = torch.stack(policy_loss).sum()
            
            # 优化模型
            self.optimizer.zero_grad()
            policy_loss.backward()
            self.optimizer.step()
        
        # 清空记忆
        self.log_probs = []
        self.rewards = []

# 训练函数
def train(agent, env, num_episodes=1000, render_every=100):
    episode_rewards = []
    
    for episode in range(num_episodes):
        state, _ = env.reset()
        episode_reward = 0
        done = False
        truncated = False
        
        # 如果是渲染回合，创建一个带渲染的环境
#         if episode % render_every == 0:
#             render_env = gym.make("FlappyBird-v0", render_mode="human", use_lidar=False)
#             render_state, _ = render_env.reset()
#             render_done = False
#             render_truncated = False
        
        while not done and not truncated:
            # 选择动作
            action = agent.select_action(state)
            
            # 执行动作
            next_state, reward, done, truncated, _ = env.step(action)
            
            # 存储奖励
            agent.store_reward(reward)
            
#             # 渲染环境（如果是渲染回合）
#             if episode % render_every == 0 and not render_done and not render_truncated:
#                 render_action = agent.select_action(render_state)
#                 render_state, _, render_done, render_truncated, _ = render_env.step(render_action)
            
            # 更新状态和累积奖励
            state = next_state
            episode_reward += reward
        
        # 学习
        agent.learn()
        
        # 关闭渲染环境
#         if episode % render_every == 0:
#             render_env.close()
        
        # 记录回合奖励
        episode_rewards.append(episode_reward)
        
        # 打印进度
        if episode % 10 == 0:
            avg_reward = np.mean(episode_rewards[-10:])
            print(f"Episode {episode}, Average Reward (last 10): {avg_reward:.2f}")
    
    return episode_rewards

# 创建智能体
input_dim = env.observation_space.shape[0]  # 观察空间维度
output_dim = env.action_space.n  # 动作空间维度
agent = PolicyGradientAgent(input_dim, output_dim)

# 训练智能体
rewards = train(agent, env, num_episodes=500000, render_every=50)

# 绘制学习曲线
plt.figure(figsize=(10, 6))
plt.plot(rewards)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Policy Gradient Learning Curve')
plt.savefig('learning_curve.png')
plt.show()

# 关闭环境
env.close()

# 保存模型
torch.save(agent.policy.state_dict(), 'flappy_bird_policy.pth')

# 测试训练好的模型
def test_agent(agent, env, num_episodes=5):
    for episode in range(num_episodes):
        state, _ = env.reset()
        done = False
        truncated = False
        total_reward = 0
        
        while not done and not truncated:
            action = agent.select_action(state)
            state, reward, done, truncated, _ = env.step(action)
            total_reward += reward
        
        print(f"Test Episode {episode+1}, Total Reward: {total_reward}")

# 创建测试环境（带渲染）
test_env = gym.make("FlappyBird-v0", render_mode="human", use_lidar=False)
test_agent(agent, test_env, num_episodes=3)
test_env.close()
```
