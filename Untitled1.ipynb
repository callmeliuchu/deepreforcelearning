```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def policy_forward(x):
    h = np.dot(model['W1'],x) ## 200
    h[h<0] = 0
    logit = np.dot(model['W2'],h)
    p = sigmoid(logit)
    return p,h
```

```python
def policy_backward(logps,h,x):
    dW2 = np.dot(h.T,logps).ravel()
    assert dW2.shape == model['W2'].shape, str(f"{dW2.shape} ==> {model['W2'].shape}")
    t = np.outer(logps,model['W2'])
    assert t.shape == h.shape
    t[h<0] = 0
    dW1 = np.dot(t.T,x)
    assert dW1.shape == model['W1'].shape
    return {'W1':dW1,'W2':dW2}
```

```python
def discount_rewards(rws,gamma=0.9):
    adding = 0
    for i in reversed(range(len(rws))):
        if rws[i] == 0:
            adding = 0
        adding = adding * gamma + rws[i]
        rws[i] = adding
    return np.array(rws)
```

```python
rewards = []

model = {
    'W1': np.random.randn(H,D),
    'W2': np.random.randn(H)
}


xs = []
hs = []
logps = []
before_x = None
reward_sum = 0
episode = 0
batch_size = 64
grad_sum = {}
decay = 0.9
learning = 0.001
rms_cache = {'W1':np.zeros_like(model['W1']),'W2':np.zeros_like(model['W2'])}
observation,_ = env.reset()
for i in range(1000000):
    pre = prepro(observation).astype(np.float32).ravel() # 6400
    cur_x = np.zeros(D) if before_x is None else pre - before_x
    before_x = pre
    p,h = policy_forward(cur_x)
    # print('cur_x',cur_x,p,h)
    xs.append(cur_x)
    hs.append(h)
    action = 2 if np.random.uniform() < p else 3
    observation,reward,terminated,truncated,info = env.step(action)
    y = 0 if action == 2 else 1
    logps.append(y-p)
    reward_sum += reward
    rewards.append(reward_sum)
    
    if terminated or truncated:
        episode += 1
        ####### 
        epx = np.vstack(xs)
        eph = np.vstack(hs)
        eplogs = np.vstack(logps).ravel()
        dis_rewards = discount_rewards(rewards)
        dis_rewards -= dis_rewards.mean()
        dis_rewards /= dis_rewards.std()
        # print('eplogs shape, dis_rewards shape',eplogs.shape,dis_rewards.shape)
        eplogs *= dis_rewards
        grad = policy_backward(eplogs,eph,epx)

        # print('xxxx',logps,eplogs,eph,epx)
        for k in grad:
            if k not in grad_sum:
                grad_sum[k] = grad[k]
            else:
                grad_sum[k] += grad[k]

        if reward_sum != 0:
            print('reward.......',reward_sum,grad_sum['W1'][:100])
                
        ## policy
        observation,_ = env.reset()
        before_x = None
        reward_sum = 0
        rewards = []
        xs = []
        hs = []
        logps = []
        
        if episode % batch_size == 0:
            ### 
            for k in grad_sum:
                rms_cache[k] = decay * rms_cache[k] + (1-decay) * grad_sum[k] ** 2
                model[k] += grad_sum[k] * learning / (1e-5 + np.sqrt(rms_cache[k]))
            grad_sum = {}
```
